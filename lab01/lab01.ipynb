## Question 1

In question 1, **Contrast Stretching** is applied. It should be noticed that **Contrast Stretching** normalizes the image.

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
img_I = cv2.imread('cat.png')
img_I = cv2.cvtColor(img_I, cv2.COLOR_BGR2RGB) # convert RGB order for plt.imshow() later
```

```python
def contrast_stretching(img_I):
    c = np.amin(img_I) # minimum pixel value occuring in image Input
    d = np.amax(img_I) # maximum pixel value occuring in image Input
    a = 0 # minimum pixel value allowed for an input image
    b = 255 # minimum pixel value allowed for an input image
    img_O = (img_I-c*np.ones(img_I.shape,np.uint8))*((b-a)/(d-c))+a 
    return (img_O)
    
# Show image
def show_img(img):
    fig = plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
    img_plot = plt.imshow(img.astype('uint8'))
    plt.show()
```

```python
img_O = contrast_stretching(img_I) 
img_O = cv2.normalize(img_O,None,0,255,norm_type=cv2.NORM_MINMAX)
cv2.imwrite('q1.png',img_O) 
show_img(img_O)
```

## Question 3

In question 3, we apply Sobel filters encapsulated in *opencv* package and self-defined Sobel filters to the original image respectively. It could be seen that, the output of the self-defined Sobel filters is the same as the *opencv* one.

### Using Sobel from opencv

```python
grad_x = cv2.Sobel(img_I,cv2.CV_16S,1,0,ksize=3)
grad_y = cv2.Sobel(img_I,cv2.CV_16S,0,1,ksize=3)
abs_grad_x = cv2.convertScaleAbs(grad_x) 
abs_grad_y = cv2.convertScaleAbs(grad_y) 
grad = cv2.addWeighted(abs_grad_x,0.5,abs_grad_y,0.5,0) 
cv2.imwrite('grad_sobel.png',grad)
show_img(grad)
```

### Without Using Sobel in Opencv

```python
sobel_x = np.array((
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]),dtype='int') 

sobel_y = np.array((
    [-1,-2,-1],
    [0,0,0],
    [1,2,1]),dtype='int') 
```

```python
grad_x = cv2.filter2D(img_I, -1, sobel_x)
grad_y = cv2.filter2D(img_I, -1, sobel_y)
```

```python
abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv2.convertScaleAbs(grad_y)
grad = cv2.addWeighted(abs_grad_x,0.5,abs_grad_y,0.5,0)
cv2.imwrite('grad.png',grad)
show_img(grad)
```

## Question 4

In question 4, **Image Sharpening** is applied to the input image.

```python
SIGMA = 1.0 
A = 1.25
```

```python
# sharpen an image with a kernel of size N in GaussianBlur()
def img_sharpening(img_I,N):
    cs_img_I = contrast_stretching(img_I)
    img_L = cv2.GaussianBlur(cs_img_I,(N,N),sigmaX=SIGMA,sigmaY=SIGMA)
    imgH = np.zeros(cs_img_I.shape,np.uint8)
    img_H = cs_img_I - img_L
    img_O = cs_img_I + A*img_H 
    return (contrast_stretching(img_O))
```

First, we have a view on when $N=0$, namely GaussianBlur() function gets the $N$ value is computed from $\sigma$ value.

```python
show_img(img_sharpening(img_I,0))
```

It could be seen that, some details in the output image are still blurry. Hence, we should apply small values $N$ to the img_sharpening() function to get the optimal output. In the following step, we try the $N$ value from 1 to 17 with step of 2.

```python
rows = 3
cols = 3
fig = plt.figure(figsize=(18, 15), dpi= 80, facecolor='w', edgecolor='k')
axes = []
for i in range(1,rows*cols*2,2):
    axes.append(fig.add_subplot(rows,cols,(i+1)//2))
    subplot_title = ("kernel size: "+str(i))
    axes[-1].set_title(subplot_title)
    plt.imshow(img_sharpening(img_I,i).astype('uint8'))
fig.tight_layout()
plt.show()
```

Since in Gaussian filter, only the value of $\sigma$ affects the smoothing effect and increasing the kernel's size does not contribute anything, from the result above, it could be seen that when $N=5$, the kernel size is large enough to cover all Gaussian effects as the images with bigger $N$ values do not have obvious quality enhancement.

```python
cv2.imwrite('sharpened_cat.png',img_sharpening(img_I,5))
```
